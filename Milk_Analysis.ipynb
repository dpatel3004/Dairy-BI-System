**KPI calculation**

```python
# Merge sales + cost
df = sales.merge(product_costs, on="product_name")

df["date"] = pd.to_datetime(df["date"], dayfirst=True, errors="coerce")
spoilage["date"] = pd.to_datetime(spoilage["date"], dayfirst=True, errors="coerce")

# Calculate cost & gross profit
df["total_cost"] = df["units_sold"] * df["total_cost_per_unit"]
df["gross_profit"] = df["revenue"] - df["total_cost"]

# Merge spoilage by date
spoilage_loss = spoilage.groupby("date")["loss_amount_rs"].sum().reset_index()

df = df.merge(spoilage_loss, on="date", how="left")
df["loss_amount_rs"] = df["loss_amount_rs"].fillna(0)

df["true_net_profit"] = df["gross_profit"] - df["loss_amount_rs"]

kpi = df["true_net_profit"].sum()

print("True Net Profit After Spoilage:", round(kpi,2))
```

```python

# -----------------------------
# 2️⃣ Standardize Date Columns
# -----------------------------
sales["date"] = pd.to_datetime(sales["date"], dayfirst=True, errors="coerce")
milk["date"] = pd.to_datetime(milk["date"], dayfirst=True, errors="coerce")
spoilage["date"] = pd.to_datetime(spoilage["date"], dayfirst=True, errors="coerce")

# -----------------------------
# 3️⃣ Merge Sales + Product Costs
# -----------------------------
df = sales.merge(product_costs, on="product_name", how="left")

# -----------------------------
# 4️⃣ Calculate Cost & Gross Profit
# -----------------------------
df["total_cost"] = df["units_sold"] * df["total_cost_per_unit"]
df["gross_profit"] = df["revenue"] - df["total_cost"]

# -----------------------------
# 5️⃣ Aggregate Sales to Daily Level
# -----------------------------
sales_daily = df.groupby("date").agg(
    total_revenue=("revenue", "sum"),
    total_cost=("total_cost", "sum"),
    gross_profit=("gross_profit", "sum")
).reset_index()

# -----------------------------
# 6️⃣ Aggregate Spoilage to Daily Level
# -----------------------------
spoilage_daily = spoilage.groupby("date").agg(
    total_spoilage_loss=("loss_amount_rs", "sum"),
    total_spoiled_liters=("spoiled_liters", "sum")
).reset_index()

# -----------------------------
# 7️⃣ Aggregate Milk Collection
# -----------------------------
milk_daily = milk.groupby("date").agg(
    total_liters_collected=("liters_collected", "sum"),
    avg_fat_percent=("fat_percent", "mean"),
    avg_snf_percent=("snf_percent", "mean")
).reset_index()

# -----------------------------
# 8️⃣ Merge All Daily KPIs
# -----------------------------
final_kpi = sales_daily.merge(spoilage_daily, on="date", how="left")
final_kpi = final_kpi.merge(milk_daily, on="date", how="left")

# Replace missing spoilage with 0
final_kpi["total_spoilage_loss"] = final_kpi["total_spoilage_loss"].fillna(0)
final_kpi["total_spoiled_liters"] = final_kpi["total_spoiled_liters"].fillna(0)

# -----------------------------
# 9️⃣ Calculate Final Net KPI
# -----------------------------
final_kpi["net_profit"] = final_kpi["gross_profit"] - final_kpi["total_spoilage_loss"]
final_kpi["profit_margin_pct"] = (
    final_kpi["net_profit"] / final_kpi["total_revenue"]
) * 100

# -----------------------------
# 🔟 Final Output (ONE TABLE)
# -----------------------------
print(final_kpi.head())

# If you want summary KPI:
print("\n====== OVERALL BUSINESS KPI ======")
print("Total Revenue:", final_kpi["total_revenue"].sum())
print("Total Cost:", final_kpi["total_cost"].sum())
print("Total Gross Profit:", final_kpi["gross_profit"].sum())
print("Total Spoilage Loss:", final_kpi["total_spoilage_loss"].sum())
print("Final Net Profit:", final_kpi["net_profit"].sum())
```
